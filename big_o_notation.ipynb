```python
import numpy as np
import matplotlib.pyplot as plt

import math
```

```python

x = list(range(0,5))

plt.figure(figsize=(15, 10)) # (width, height)

plt.plot(x,[n + 1 for n in x], 'y', label='$O(1)$')
plt.plot(x, [np.log(n) for n in x], 'g', label='$O(log n)$')
plt.plot(x, [n for n in x] , 'b', label='$O(n)$')
plt.plot(x, [n * (np.log(n)) for n in x], 'r', label='$O(n log n)$')
plt.plot(x,[n ** 2 for n in x], 'orange', label='$O(n^2)$')
plt.plot(x, [2 ** n for n in x], 'purple', label='$O(2^n)$')
plt.plot(x, [math.factorial(n) for n in x] , 'black', label='$O(n!)$')
plt.plot(x, [n ** n  for n in x], 'pink', label='$O(n^n)$')

plt.legend(loc='upper left')
plt.suptitle('Big O Notation',fontsize=28)
plt.xlabel('Elements')
plt.ylabel('Operations')

#hfont = {'fontname': 'Rockwell'}
hfont = {'fontname': 'Academy Engraved LET'}

plt.title('Veronica F. Reyes', loc='left', fontsize=20, **hfont)


plt.show()
```
