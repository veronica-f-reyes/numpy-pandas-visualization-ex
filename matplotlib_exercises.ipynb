### 1. Use matplotlib to plot the following equation:

    y = x ^ 2 - x + 2

    You'll need to write the code that generates the x and y points.

    Add an anotation for the point 0, 0, the origin.

```python
import numpy as np
import matplotlib.pyplot as plt

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
x = list(range(0, 50))
y = [n ** 2 - n + 2 for n in x]
plt.scatter(x, y)
plt.title('$y = x^2 - x + 2$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.annotate('Origin', xy=(0, 0), xytext=(-3, 5),
             arrowprops={'facecolor': 'blue'})
plt.show()
```

### Combine the figures you created in the last step into one large figure with 4 subplots.

```python
# some data to play with
x = list(range(0, 50))


# plot the first subplot
plt.subplot(221)
plt.plot(x,[np.sqrt([n]) for n in x] )
plt.title(r'$\sqrt{2}$')

# the second subplot
plt.subplot(222)
plt.plot(x, [n ** 3 for n in x])
plt.title('$y = x^3$')

plt.subplot(223)
plt.plot(x, [2 ** n for n in x])
plt.title('$y = 2^x$')

plt.subplot(224)
plt.plot(x, [1/(n+1) for n in x])
plt.title('$y = 1 / (x+1)$')

plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.45,
                    wspace=0.35)


plt.show()
```

### Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
# some data to play with
x = list(range(0, 7))

plt.figure(figsize=(14, 10)) # (width, height)

plt.plot(x,[np.sqrt([n]) for n in x], 'y', label='y = $\sqrt{x}$')
plt.plot(x, [n ** 3 for n in x], 'g', label='$y = x^3$')
plt.plot(x, [2 ** n for n in x] , 'b', label='$y = 2^x$')
plt.plot(x, [1/(n+1) for n in x], 'r', label='$y = 1 / (x+1)$')

plt.legend(loc='upper right')
plt.title('Approximating curves')
plt.xlabel('$x$')
plt.ylabel('$y$')


plt.show()
```
